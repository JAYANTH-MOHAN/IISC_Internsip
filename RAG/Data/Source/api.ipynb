```python
import requests
import pandas as pd
import newsapi
from newsapi import NewsApiClient
import os


def get_articles_by_date(api_key, keyword, from_date, to_date, sort_by='popularity', page_size=100, page=1):
    base_url = "https://newsapi.org/v2/everything"

    params = {
        'q': keyword,
        'apiKey': api_key,
        'from': from_date,
        'to': to_date,
        'sortBy': sort_by,
        'pageSize': page_size,
        'page': page,
        'language': 'en'  # Add the language parameter for English
    }

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        return response.json()['articles']
    else:
        print(f"Error: {response.status_code}")
        return None

# Replace 'your_api_key' with your actual News API key
api_key = 'ab9b8f8aa73749209313ff8c3c7a60aa'
#ab9b8f8aa73749209313ff8c3c7a60aa
#78ba152e740449aca9b8184bb4777f33
# Replace 'your_keyword' with the keyword you want to search for (in this case, 'apple')
keyword = 'Israel Hamas war'

# Specify the date in the format 'YYYY-MM-DD'
from_date = '2024-02-09'
to_date = '2024-02-14'
# Get articles
articles = get_articles_by_date(api_key, keyword, from_date, to_date)

# Specify the CSV file name
csv_file = 'articles_data_14_2_3.csv'

# Check if the CSV file already exists
if os.path.isfile(csv_file):
    # Load existing DataFrame from the CSV file
    existing_df = pd.read_csv(csv_file)

    # Append new data to the existing DataFrame
    if articles:
        new_df = pd.DataFrame(articles)
        updated_df = pd.concat([existing_df, new_df], ignore_index=True)

        # Save the updated DataFrame to the CSV file
        updated_df.to_csv(csv_file, index=False)

        # Print the updated DataFrame
        print(updated_df)
        print("Data appended to 'articles_data.csv'")
    else:
        print("No new articles found.")
else:
    # If the CSV file doesn't exist, create a new one
    if articles:
        df = pd.DataFrame(articles)
        df.to_csv(csv_file, index=False)
        print(df)
        print("Data saved as 'articles_data.csv'")
    else:
        print("No articles found.")
```
